```python
from itertools import chain
class myiterabledataset_(torch.utils.data.IterableDataset):
    def __init__(self,data_list):
        self.data_list=data_list
    def parse_file(self,file_path):
        with open(file_path,'r') as file_obj:
            for line in file_obj:
                tokens=line
                yield from tokens
    def get_stream(self):
        return chain.from_iterable(map(self.parse_file,self.data_list))
    def __iter__(self):
        return iter(self.get_stream())
```

```python
myiterabledataset=myiterabledataset_(["win2.xls","wine1.xls"])
```

```python
loader=torch.utils.data.DataLoader(myiterabledataset,batch_size=5)
print(loader)
for i,batch in enumerate(loader):
    print('num_of_batch is:',i)
    print(batch)
```
